```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split

# Load AG News dataset (adjust path & column names if needed)
df = pd.read_csv("../data/ag_news_train.csv")   # columns: label, title, description

# Merge title + description into one text column
df['text'] = (df['title'].fillna('') + " " + df['description'].fillna('')).str.lower()

# Basic cleaning: remove links and extra spaces
df['text'] = df['text'].apply(lambda x: re.sub(r"http\S+|www\.\S+", " ", str(x)))
df['text'] = df['text'].str.replace(r"\s+", " ", regex=True).str.strip()

# Split into train/val
X_train, X_val, y_train, y_val = train_test_split(
    df['text'], df['label'], test_size=0.16666, stratify=df['label'], random_state=42
)

print("Train size:", len(X_train), "Validation size:", len(X_val))
print("Classes:", df['label'].unique())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
# ---- Pipeline: TF-IDF -> χ² SelectKBest -> Logistic Regression ----
K = 30000  # tune if needed (e.g., 10k–60k depending on RAM)
pipe_fs = Pipeline([
    ('tfidf', TfidfVectorizer(
        max_features=120_000,
        ngram_range=(1,2),
        stop_words='english',
        lowercase=True
    )),
    ('kbest', SelectKBest(chi2, k=K)),
    ('clf', LogisticRegression(max_iter=300, n_jobs=-1))
])

pipe_fs.fit(X_train, y_train)
pred_fs = pipe_fs.predict(X_val)

print("== χ² Feature Selection Report ==")
print(classification_report(y_val, pred_fs))
```

```python
# ---- Visualization: Top-20 χ² features by score (overall) ----
tfidf = pipe_fs.named_steps['tfidf']
kbest = pipe_fs.named_steps['kbest']

feature_names = np.array(tfidf.get_feature_names_out())
scores = kbest.scores_

# scores is aligned to tfidf features BEFORE selection
# Get top 20 feature indices by score
top_idx = np.argsort(scores)[-20:]
top_feats = feature_names[top_idx]
top_scores = scores[top_idx]

# Sort for prettier plot
order = np.argsort(top_scores)
top_feats = top_feats[order]
top_scores = top_scores[order]

plt.figure(figsize=(8,6))
plt.barh(top_feats, top_scores)
plt.title("Top 20 χ²-selected features (overall)")
plt.xlabel("χ² score")
plt.tight_layout()
plt.show()
```

```python
# ---- (Optional) Confusion matrix to visualize performance ----
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_val, pred_fs)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(xticks_rotation=0)
plt.title("Confusion Matrix — χ² + KBest")
plt.tight_layout()
plt.show()
```
