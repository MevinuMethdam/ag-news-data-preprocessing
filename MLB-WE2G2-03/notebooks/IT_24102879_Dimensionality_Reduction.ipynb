```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split

# Load AG News dataset (adjust path & column names if needed)
df = pd.read_csv("../data/ag_news_train.csv")   # columns: label, title, description

# Merge title + description into one text column
df['text'] = (df['title'].fillna('') + " " + df['description'].fillna('')).str.lower()

# Basic cleaning: remove links and extra spaces
df['text'] = df['text'].apply(lambda x: re.sub(r"http\S+|www\.\S+", " ", str(x)))
df['text'] = df['text'].str.replace(r"\s+", " ", regex=True).str.strip()

# Split into train/val
X_train, X_val, y_train, y_val = train_test_split(
    df['text'], df['label'], test_size=0.16666, stratify=df['label'], random_state=42
)

print("Train size:", len(X_train), "Validation size:", len(X_val))
print("Classes:", df['label'].unique())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
```

```python
# ---- Pipeline: TF-IDF -> TruncatedSVD (LSA) -> Logistic Regression ----
n_components = 300  # common choices: 100â€“500
pipe_svd = Pipeline([
    ('tfidf', TfidfVectorizer(
        max_features=120_000,
        ngram_range=(1,2),
        stop_words='english',
        lowercase=True
    )),
    ('svd', TruncatedSVD(n_components=n_components, random_state=42)),
    ('clf', LogisticRegression(max_iter=300, n_jobs=-1))
])

pipe_svd.fit(X_train, y_train)
pred_svd = pipe_svd.predict(X_val)

print("== LSA (TruncatedSVD) Report ==")
print(classification_report(y_val, pred_svd))
```

```python
# ---- Visualization 1: Cumulative explained variance ----
svd_model = pipe_svd.named_steps['svd']
cumvar = np.cumsum(svd_model.explained_variance_ratio_)

plt.figure(figsize=(7,5))
plt.plot(range(1, n_components+1), cumvar)
plt.title("Cumulative explained variance â€” TruncatedSVD")
plt.xlabel("Number of components")
plt.ylabel("Explained variance ratio")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Total explained variance with {n_components} comps: {cumvar[-1]:.3f}")
```

```python
# ---- Visualization 2 (optional): 2D scatter of first two components ----
# Take a small sample for clarity
sample_idx = np.random.RandomState(42).choice(len(X_val), size=min(1500, len(X_val)), replace=False)
# Transform X_val to 2D
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_only = pipe_svd.named_steps['tfidf']
X_val_tfidf = tfidf_only.transform(X_val.iloc[sample_idx])
X_val_2d = TruncatedSVD(n_components=2, random_state=42).fit_transform(X_val_tfidf)

labels_sample = np.array(y_val.iloc[sample_idx])

plt.figure(figsize=(7,6))
for lab in np.unique(labels_sample):
    mask = labels_sample == lab
    plt.scatter(X_val_2d[mask, 0], X_val_2d[mask, 1], s=10, alpha=0.6, label=str(lab))
plt.legend(title="Class")
plt.title("Val set projected to 2D (SVD on TF-IDF)")
plt.xlabel("Component 1"); plt.ylabel("Component 2")
plt.tight_layout()
plt.show()
```

```python
# ---- Confusion matrix to visualize performance ----
cm = confusion_matrix(y_val, pred_svd)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(xticks_rotation=0)
plt.title("Confusion Matrix â€” LSA (SVD)")
plt.tight_layout()
plt.show()
```
